```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Helper function to display multiple images in a plot
def display_images(images, rows, cols, size, title, subtitles):
    fig = plt.figure(figsize = size)
    c = 0
    sz = len(images)
    for i in range(rows):
        for j in range(cols):
            if sz == c:
                break
            ax1 = fig.add_subplot(rows, cols, c + 1)
            ax1.imshow(images[c], cmap = 'gray')
            ax1.set_title(subtitles[c], fontsize = 20)
            c = c + 1
    plt.show()
    # Un comment the below line and comment the above line to save and display
    #plt.savefig('part2.png')
```

### Function generating a video from the frames

```python
def framesToVideo(frame_folder, video_path, fps):
    if frame_folder.endswith('/'):
        frame_folder = frame_folder[0: len(frame_folder) - 1]
    files = os.listdir(frame_folder)
    files.sort()
    frame_size = -1
    output = -1
    for f in files:
        if f.endswith('.png') or f.endswith('.jpg') or f.endswith('.jpeg'):
            frame = cv2.imread(frame_folder + '/' + f)
            if frame_size == -1:
                height, width, channels = frame.shape
                frame_size = (width, height)
                print(frame_size)
                output = cv2.VideoWriter(video_path, cv2.VideoWriter_fourcc(*'DIVX'), fps, frame_size)
            output.write(frame)
    output.release()
```

### Functions extracting frames from a video and saving them in a particular folder

```python
def frameName(c, m):
    cnt = 0
    cpy = c
    while c > 0:
        c = c // 10
        cnt = cnt + 1
    res = "frame"
    for i in range(m - cnt):
        res = res + "0"
    res = res + str(cpy)
    return res + ".jpg"

def videoToFrame(video_path, frame_folder):
    if frame_folder.endswith('/'):
        frame_folder = frame_folder[0: len(frame_folder) - 1]
    video = cv2.VideoCapture(video_path)
    success, image = video.read()
    count = 1
    while success:
        cv2.imwrite(frame_folder + '/' + frameName(count, 4), image)
        success, image = video.read()
        count = count + 1
```

### Function capturing frames from a web cam and saving them at a particular location

```python
def captureVideo(frame_folder):
    if frame_folder.endswith('/'):
        frame_folder = frame_folder[0: len(frame_folder) - 1]
    cap = cv2.VideoCapture(0)
    count = 1
    while(True):
        ret, frame = cap.read()
        cv2.imshow('frame', frame)
        cv2.imwrite(frame_folder + '/' + frameName(count, 4), frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        count = count + 1
    cap.release()
    cv2.destroyAllWindows()
```

### Functions implementing Chroma Keying on images

```python
def most_frequent_color(im):
    if len(im.shape) == 3:
        u, c = np.unique(im.reshape([1, -1, 3]), axis = 1, return_counts=True)
        t = list(zip(c, u[0]))
        t = sorted(t, key = lambda x : x[0])
        return t[len(t) - 1][1]
    else:
        print("Grayscale image doesn't make sense here!")

def mergeImage(fg, bg, threshold):
    color = most_frequent_color(fg)
    fg = fg.astype(np.float64)
    bg = bg.astype(np.float64)
    img = np.ones(fg.shape)    
    img = np.multiply(np.dstack([np.sum(abs(fg - color), axis = 2) <= threshold] * 3), bg) + np.multiply(np.dstack([np.sum(abs(fg - color), axis = 2) > threshold] * 3), fg)
    return img.astype(np.uint8)
```

### Function implementing chroma keying for videos

```python
def ChromaKeyVideo(fg_video_path, bg_video_path, video_path, fps):
    fg_video = cv2.VideoCapture(fg_video_path)
    bg_video = cv2.VideoCapture(bg_video_path)
    success1, fg_image = fg_video.read()
    success2, bg_image = bg_video.read()
    count = 1
    output = -1
    while success1 and success2:
        height, width, channels = fg_image.shape
        frame_size = (width, height)
        if output == -1:
            output = cv2.VideoWriter(video_path, cv2.VideoWriter_fourcc(*'DIVX'), fps, frame_size)
        img = mergeImage(fg_image, bg_image, 110)
        output.write(img)
        success1, fg_image = fg_video.read()
        success2, bg_image = bg_video.read()
        count = count + 1
    output.release()
```

### Function implementing Viola Jones face detector for web cam capture

```python
def detect_face(img):
    face_cascade = cv2.CascadeClassifier('../files/haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
    return img
def face_detection_capture(video_folder, fps):
    if video_folder.endswith('/'):
        video_folder = video_folder[0: len(video_folder) - 1]
    cap = cv2.VideoCapture(0)
    count = 1
    output = -1
    while(True):
        ret, frame = cap.read()
        frame = detect_face(frame)
        cv2.imshow('frame', frame)
        if count == 1:
            height, width, channels = frame.shape
            frame_size = (width, height)
            output = cv2.VideoWriter(video_folder, cv2.VideoWriter_fourcc(*'DIVX'), fps, frame_size)
        output.write(frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        count = count + 1
    cap.release()
    output.release()
    cv2.destroyAllWindows()
```
